Shuffle Phase

```python
from collections import defaultdict

def shuffle(mapped_data):
    grouped = defaultdict(list)
    for key, value in mapped_data:
        grouped[key].append(value)
    return grouped
```

Reducer Phase

```python
from collections import defaultdict

def reducer(mapped_data):
    grouped = defaultdict(int)
    for key, value in mapped_data:
        grouped[key] += value
    return grouped
```

**Bonus 1: عدد الطلبات لكل URL**

```python
def mapper_url(line):
    fields = line.strip().split(',')
    if len(fields) != 7 or fields[0].startswith('#'):
        return []
    url = fields[4]       # الحقل الخامس هو URL
    return [(url, 1)]
```

```python
# -----------------------------------
# 1. Mapper
# -----------------------------------
def mapper(line):
    try:
        timestamp, ip, method, endpoint, code, size = line.strip().split(",")
        return [(code, 1)]
    except:
        return []


# -----------------------------------
# 2. Reducer
# -----------------------------------
def reducer(mapped_list):
    reduced = {}
    for code, count in mapped_list:
        reduced[code] = reduced.get(code, 0) + count
    return reduced


# -----------------------------------
# 3. Main program
# -----------------------------------
mapped = []

with open("weblogs.txt", "r") as f:
    for line in f:
        mapped.extend(mapper(line))

# reduce
reduced = reducer(mapped)

# print results
for code, count in sorted(reduced.items()):
    print(f"HTTP {code}: {count} requests")

```

**Bonus 2: مجموع ResponseSize لكل Status Code**

```python
def mapper_size_per_status(line):
    fields = line.strip().split(',')
    if len(fields) != 7 or fields[0].startswith('#'):
        return []
    status = fields[5]
    size = int(fields[6])   # ResponseSize
    return [(status, size)]
```

```python
from collections import defaultdict

def reducer_sum(mapped_data):
    grouped = defaultdict(int)
    for key, value in mapped_data:
        grouped[key] += value
    return grouped
```

```python
mapped_size = []
with open("weblogs.txt", "r") as f:
    for line in f:
        mapped_size.extend(mapper_size_per_status(line))

reduced_size = reducer_sum(mapped_size)

print("Total Response Size per Status Code:")
for code, total_size in sorted(reduced_size.items()):
    print(f"HTTP {code}: {total_size} bytes")
```

**Bonus 3: تحليل الأخطاء فقط (بدون 200)**

```python
def mapper_errors_only(line):
    fields = line.strip().split(',')
    if len(fields) != 7 or fields[0].startswith('#'):
        return []
    status = fields[5]
    if status == '200':
        return []          # تجاهل الطلبات الناجحة
    return [(status, 1)]
```

```python
mapped_errors = []
with open("weblogs.txt", "r") as f:
    for line in f:
        mapped_errors.extend(mapper_errors_only(line))

reduced_errors = reducer(mapped_errors)

print("Error Requests (non-200):")
for code, count in sorted(reduced_errors.items()):
    print(f"HTTP {code}: {count} requests")
```
